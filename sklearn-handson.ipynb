### Visual EDA

### Build classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(iris['data'], iris['target'])
```

```python
iris['data'].shape
```

```python
iris['target'].shape
```

### Do the prediction!

```python
X_new = np.array([[5.2, 3.2, 4.2, 1.2],[1.2, 2.2, 3.2, 7.2]])
prediction = knn.predict(X_new)
prediction      
```
